```python
# Importa a biblioteca pandas, e lê o arquivo excel importado do ranking IPS

import pandas as pd

# 1. Ler a planilha
df = pd.read_excel('/content/ips_brasil_municipios_2025_rj.xlsx')

# 2. Mostrar as primeiras linhas
df.head()
```

```python
# Separa e filtra a planilha pela UF do Rio

df_rj = df[df['UF'] == 'RJ']

print(f'Total de municípios do RJ: {len(df_rj)}')

df_rj.head()
```

```python
# Separa as colunas da tabela que eu julgo relevante para o projeto

colunas = [
    'Município',
    'PIB per capita 2021',
    'Cobertura Vacinal (Poliomielite)',
    'Expectativa de Vida',
    'Mortalidade Infantil até 5 Anos',
    'Hospitalizações por Condições Sensíveis à Atenção Primária',
    'Acesso ao Conhecimento Básico',
    'Acesso à Educação Superior',
    'Nota Mediana do Enem',
    'Assassinatos de Jovens',
    'Assassinatos de Mulheres',
    'Homicídios',
    'Abastecimento de Água via Rede de Distribuição',
    'Esgotamento Sanitário Adequado',
    'Domicílios com Coleta de Resíduos Adequada',
    'Domicílios com Iluminação Elétrica Adequada',
    'Áreas Verdes Urbanas',
    'Emissões de CO₂e por Habitante',
    'Índice de Vulnerabilidade Climática dos Municípios (IVCM)',
    'Acesso a Programas de Direitos Humanos',
    'Paridade de Gênero na Câmara Municipal',
    'Paridade de Negros na Câmara Municipal',
    'Inclusão Social'
]

df_rj = df_rj[colunas]
df_rj.head()
```

```python
# Converter tudo que for número para tipo numérico (corrige vírgulas, textos etc.)
for col in colunas[1:]:
  df_rj[col] = pd.to_numeric(df_rj[col], errors='coerce')

# Analisa e Substitui valores vazios pela média da coluna
df_rj.fillna(df_rj.mean(numeric_only=True), inplace=True)

# Normaliza os dados, colocando todos os dados entre 0 e 1 *
for col in colunas[1:]:
    df_rj[col] = (df_rj[col] - df_rj[col].min()) / (df_rj[col].max() - df_rj[col].min())



# * -> Isso é importante, porque o PIB, por exemplo, tem valores muito maiores que a cobertura vacinal
# e a normalização impede que ele domine o ranking.
```

```python
# Arredondar todos os valores numéricos para 2 casas decimais

df_rj = df_rj.round(2)
```

```python
# Salva a base tratada e gera um arquivo .csv chamado 'dados_limpos_rj.csv'

df_rj.to_csv('/content/dados_limpos_rj.csv', index=False)
print("Base de dados tratada e salva com sucesso! ✅")
```
